```python
import os
from bs4 import BeautifulSoup as bs
import pandas

farming=[]
biota=[]
boundaries=[]
climatologyMeteorologyAtmosphere=[]
economy=[]
elevation=[]
environment=[]
geoscientificInformation=[]
health=[]
imageryBaseMapsEarthCover=[]
intelligenceMilitary=[]
inlandWaters=[]
location=[]
oceans=[]
planningCadastre=[]
society=[]
structure=[]
transportation=[]
utilitiesCommunication=[]


concordance = {
    "farming":farming,
    "biota":biota,
    "boundaries":boundaries,
    "climatologyMeteorologyAtmosphere":climatologyMeteorologyAtmosphere,
    "economy": economy,
    "elevation": elevation,
    "environment":environment,
    "geoscientificInformation":geoscientificInformation,
    "health":health,
    "imageryBaseMapsEarthCover":imageryBaseMapsEarthCover,
    "intelligenceMilitary":intelligenceMilitary,
    "inlandWaters":inlandWaters,
    "location":location,
    "oceans":oceans,
    "planningCadastre":planningCadastre,
    "society":society,
    "structure":structure,
    "transportation":transportation,
    "utilitiesCommunication":utilitiesCommunication
}

# filenames "fgcd12345.xml" from "1" to "10016" but missing "9985"-"9999"

baseFilePath = "/Documents/FGDC_xml/fgdc"
filerangeStart = 1
filerangeEnd = 10016
fileCount = 0
listByFile = []

beginMsg = "Beginning with file number {} "
print(beginMsg.format(filerangeStart))

for i in range (filerangeStart, filerangeEnd):
    filename = baseFilePath + str(i) + ".xml"
    if os.path.isfile(filename):
        isoThemes = set()
        lcshThemes = set()
        themesList = []

        with open (filename, "r", encoding="utf-8") as f:
            themeString = ""
            fgdc = bs(f, "xml")
            keywords = fgdc.find_all("keywords")

            #Search XML for <theme> in each record
            for keyword in keywords:
                themes = keyword.find_all("theme")
                for theme in themes:
                    #isolate the type of theme
                    themeString = theme.find_all()[0].string
                    if(themeString == "ISO 19115 Topic Category"):
                        themekeys = theme.find_all("themekey")
                        for themekey in themekeys:
                            isoThemes.add(themekey.string)
                        #print(filename, isoThemes)
                    elif (themeString == "LCSH"):
                        #print(themeString, "===>")
                        themekeys = theme.find_all("themekey")
                        for themekey in themekeys:
                            lcshThemes.add(themekey.string)

            # After processing each xml file, add key to concordance if not already present, 
            # (to catch erroneous terms) and append lcsh set
            for isotheme in isoThemes:
                themesList.push(isotheme) #tracking by file, to get frequency count information
                if not(isotheme in concordance):
                    concordance[isotheme] = []
                for lcsh in lcshThemes:
                    concordance[isotheme].append(lcsh)
                    themesList.push(lcsh)
            listByFile.push(themesList) #list of lists, one per file.
            fileCount += 1
            if fileCount%250 == 0:
                print("Processed",fileCount, "files...")
    else:
        print(filename, "not found.")

isoLcsh = {}

for k, v in concordance.items():
    if len(v) > 0:
        isoLcsh[k]=list(set(v))

# Create dataframe; creating it with orient='index' (default is the other choice, 'column')
# and then transposing is one trick for handling errors arising from columns of unequal length. 

#df = pandas.DataFrame.from_dict(isoLcsh, orient='index')
#dft = df.transpose()

endMessage = ("Processed {} files")
print(endMessage.format(fileCount))

#export to csv

#csv_path = r"iso_lcsh_themes.csv"

#export_csv = dft.to_csv (csv_path, index = None, header=True)
#print("Exporting Data Frame results to:", csv_path)

##pp = pprint.PrettyPrinter(indent=4)
##pp.pprint(isoLcsh)

# Write the big 'list by file' to text; figure out what to do with it...
with open ('themeListByFile.txt', 'w') as l:
    for item in listByFile:
        t.write("%s\n" % item)


     
```
